```python
# Import Libraries

import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the figuration of the plots

# Read data

df = pd.read_csv(r'C:\Users\jamie\Downloads\movies.csv')
```

```python
df.head()
```

```python
# Check for missing data

for col in df.columns: 
    missing_pct = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, missing_pct))
```

```python
# Data types of columns

df.dtypes
```

```python
# Convert budget and gross data type to integer

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Create new released year column using the correct year

df['releasedyear'] = df['released'].astype(str).str[:4]

df
```

```python
df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# Remove duplicates

df.drop_duplicates()
```

```python
# Scatter plot with budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Film Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')

plt.show()
```

```python
# Plot budget vs gross vs seaborn

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color": "orange"}, line_kws = {"color": "red"})
```

```python
# Investigae correlation

df.corr()
```

```python
# Visualize correlation matrix

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot = True)

plt.title('Correlation Matrix of Movie Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
corr_pairs = corr_matrix.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr_pairs = sorted_pairs[(sorted_pairs > 0.5) & (sorted_pairs < 1.0)]

high_corr_pairs
```
